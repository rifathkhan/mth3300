# 1. File Validation

While we're talking about files: how do you make your program behave gracefully when it tries to read a file it can't find?  

We've seen *exceptions* before: when we have code that is susceptible to errors, we encase it in a try block, and if an exception (i.e., an error) is encountered, the program will rewind, and instead execute the except block.

```python
# EXAMPLE 1a: Files and exceptions
filename = input("Enter file name: ")

# Remember exceptions?  The file opening line is prone to errors, because files may not be found.  So we try it,
# and provide an alternative behavior if an error does indeed occur.
try:
    file = open(filename, "r")
    print(file.read())
    file.close()
except FileNotFoundError:
    print("No file by that name here! I'm not going to ask again because I'm lazy.")
    
print("All good!")
```

Ok, time for some real code.

```python
# EXAMPLE 2a: Ten Hundred Most Common Words

# First, it would be a good idea to read the dictionary file's words into a list
my_dictionary = open("smallwords.txt", "r")
dic_list = my_dictionary.read().split()
my_dictionary.close()

# The text!
text = """it is hard to explain a thing when you only use the ten hundred words that people use the most
i am trying to do it right now
but i am not sure if i am doing it right
i would like a computer to help me check if i am doing it right"""
text_list = text.split()

# This variable will contain the answer to the question: 
# "out of the words in the text so far, is every one in the dictionary?"
all_good = True

# Go through every word in the text
for text_word in text_list:
    found_yet = False # Have we found text_word in the dictionary yet?
    
    # For each word in the dic_list, and compare it to text_word.
    # If they are the same, found_yet = True, stop searching through the dic_list.
    for dic_word in dic_list:
        if text_word == dic_word:
            found_yet = True
            break
            
    if not found_yet:
        # If you never found text_word in the dictionary, that's a problem! So:
        all_good = False
        print("\"" + text_word + "\" is not in the dictionary!!")
        
        
# Finally, if all_good still is true, that means that our text has every word in the dictionary.
if all_good:
    print("Every word was in the dictionary.")
```

# 3. Nested Lists (or 2D Lists)

Just like we have loops within loops, we can have lists within lists.  You might have already imagined this.  In fact, shocker: if you want to store a table to a variable -- well, you can just think of that as a list of lists.

```python
# EXAMPLE 3a: A list of lists -- or, a "2-dimensional" list.

my_table = [[5,2,3,8],[4,1,2,3],[1,1,2,0,7]]

print("my_table[1][2] =", my_table[1][2])
print("my_table[1] =", my_table[1])
print("len(my_table) =", len(my_table))

my_table[0][2] += 10
print("my_table =", my_table)

```

Notice a couple of things.

* First, the lists contained in a list don't have to be the same length.  In fact, they don't even have to contain the same datatypes!  
* We access (and update) entries using two pairs of brackets.  This should be easy to understand.  `my_table[1]` is the second element of `my_table`, which is a list.  Then `(my_table[1])[2]` just means the third element of the list `my_table[1]`.  Python allows you to use those wimply parentheses, but nobody actually writes them -- this entry is simply written as `my_table[1][2]`.
* The length of `my_table` is the number of elements the list contains -- what those elements are has no bearing on this calculation.

<br><br><br><br><br><br><br><br><br><br>

Now, how can we pretty print a 2D list? If you want to do some interesting operation with a list, you use a `for` loop.  For
2D lists, you use nested `for` loops.  Let's do this together.

```python
# EXAMPLE 3b: Update a 2D List

my_table = [[5,2,3,8],[4,1,2,3],[1,1,2,0,7]]

for row in my_table:
    for elt in row:
        print("{0:5}".format(elt), end = " ")
    print("\n")
```

<br><br><br><br><br><br><br><br><br><br>

Now suppose we want to take the sum of all the entries in each row.  See if you can do that.

```python
# EXAMPLE 3c: Print out the sum of each row of my_table.

my_table = [[5,2,3,8],[4,1,2,3],[1,1,2,0,7]]

runsum = 0

for row in my_table:
    for elt in row:
        runsum += elt
    print(runsum)
    runsum = 0
    
        
```

Now, let's take `my_table`, and multiply each entry by 10.  How do we do that?  Certainly not by
`my_table = my_table*10`
-- that's legal code, but does not do anything like what we intend here.  No, we do it by using a nested loop!  

We'll need to be careful.  We've had this issue before: if you want to update, loop through *positions*, not *values*. So there are going to be some `range(len())`'s here.

```python
# EXAMPLE 3d: Update a 2D List

my_table = [[5,2,3,8],[4,1,2,3],[1,1,2,0,7]]

for i in range(len(my_table)):
    for j in range(len(my_table[i])):
        my_table[i][j] *= 10

print(my_table)
```

```python
# EXAMPLE 4a: Cat.

import matplotlib.pyplot as plt
from skimage import io

cat = io.imread("cat.jpg")

print(cat)

plt.imshow(cat)
plt.show()
```

There are two modules used in this example that we haven't seen before: matplotlib and scikit-image. We'll probably discuss at least the former down the road.  But their main purpose here is to help us illustrate that images are basically 2D lists!

<br><br><br><br><br><br><br><br><br><br>

Actually, that's a lie: they are really **3D** lists.

Python (via scikit-image) represents an image as a list of horizontal rows -- that's the first dimension.

Each row is a list of pixels -- that's the second dimension.

And each pixel is three integers -- a red value, a green value, and a blue value, which each range between 0 and 255. [Warning -- they're actually not `int`s, but rather some closely related data type.] 

```python
# EXAMPLE 4b: Cat, partially explained.

# This import gives us access to the imread function, used below.
from skimage import io

# This opens the cat file, and records its pixels as a 3D list of numbers.
cat = io.imread("cat.jpg")

# Now, let's look at a few interesting pixels. First, the top left corner, which is pretty dark.
print(cat[0][0])

#Now, how about row 25, column 390 -- that looks pretty bright.
print(cat[24][389])

#Finally, how about row 375, column 400? It's real green.
print(cat[374][399])
```

<br><br><br><br><br><br><br><br><br><br>

Now, we can play with this.  For example, watch this.

```python
# EXAMPLE 4c: Scary Cat

from skimage import io # Import for reading and converting images
import matplotlib.pyplot as plt # Import for displaying images

cat = io.imread("cat.jpg") # Function for loading


for i in range(400):
    for j in range(400):
        # For pixel (i,j), this code "flips" all the levels.  If a pixel has a level near 0, that level will now be near
        # 255, and vice-versa. What effect does this have?
        cat[i][j][0] = 255-cat[i][j][0]
        cat[i][j][1] = 255-cat[i][j][1]
        cat[i][j][2] = 255-cat[i][j][2]

# These two lines prepare a display, and then actually display it.
plt.imshow(cat)
plt.show()
```

<br><br><br><br><br><br><br><br><br><br>

Let's do one more: we'll boost red-levels, using the following rules: 
* if the red level is less than 128, we'll multiply by 2 (we don't want the red level going over 255!)
* if the red level is greater than or equal to 128, we'll just set the red level to be 255.

If that's too easy: superimpose the French flag on the cat.

```python
# EXAMPLE 4d: Red Cat

from skimage import io # Import for reading and converting images
import matplotlib.pyplot as plt # Import for displaying images

cat = io.imread("cat.jpg") # Function for loading


for i in range(400):
    for j in range(134):
        cat[i][j][2] = 255
    for j in range(268,400):
        cat[i][j][0] = 255
        
        
# These two lines prepare a display, and then actually display it.
plt.imshow(cat)
plt.show()
```

# 5. The Truth About Variables and Objects

Remember this?  What's going on here???

```python
# EXAMPLE 5a: Assignment

# Compare the results of this bit of code....
x = "Old"
y = x
x = "New"
print(y)

# ...with this one... #
x = ["Old", "One"]    #
y = x                 #   BLOCK A
x = ["New", "Guy"]    #
print(y)              #

#...AND THIS ONE!!!!! #
x = ["Old", "One"]    #
y = x                 #   BLOCK B
x[0] = "New!!!!"      #
print(y)              #
```

Why are the last two snippets so different?  They both change `x`, but one change causes `y` to change, and the other doesn't. 

The short answer is:

* the line `y = x` (in Block A and Block B), makes `y` and `x` refer to the same object in memory;

* the subsequent line in Block A, starting with `x = `, reassigns `x` to an *entirely new object*, while `y` still is bound to the old object;

* the subsequent line in Block B, starting with `x[0] = `, is a *mutation* which alters, but does not replace, the object which `x` and `y` both point to.  

But that probably doesn't make a lot of sense.  To make sense out of this, I need to go back to the beginning, and tell you the truth about variables, objects, and your computer's memory. 

(What follows is still not 100% accurate, but it does give a morally correct overview of how Python manages your computer's resources, and explains the above phenomenon.)

<br><br><br><br><br><br><br><br><br><br>

Let's say I ask you to multiply 5201 by 3157. You probably can't do that in your head: you need to write down those numbers, and write down the results of intermediate computations.  Maybe you write it down on the whiteboard, or on a piece of paper.

If I ask the computer to multiply 5201 by 3157, it needs to "write things down", too.  Its version of the whiteboard is the Random Access Memory, or RAM (we'll just call it the memory).  The two big differences between the computer and humans is that: the computer doesn't have eyes, so it uses numbers called *memory addresses* to keep track of where it exactly it wrote everything down; and the computer writes down only 0's and 1's, using circuitry instead of ink (we already discussed this).  I'll use `0x` to indicate that I'm writing down a memory address in the examples I show you -- these addresses will be made-up numbers, just meant to help to assist explanations, but your computer does use a numerical addressing system kind of like what I'm doing.

------------------


For every value Python needs -- every literal value it encounters, and every evaluation of an expression -- it creates an **_object_**: a parcel of memory which contains a *value* and information about what *data type* that value represents, and which has an *address*.   

I've intimated that variables are attached to values.  It's maybe more accurate to say that variables are bound to **addresses**.  In fact, Python maintains a sort-of "table", which matches defined variables to the addresses of the  objects they are bound to.

<br><br><br><br><br><br><br><br><br><br>

Let's look at a simple program to illustrate.

```python
# EXAMPLE 5b: A simple program

var = 5
x = var + 1
var = x
print(var)
```

Here's what Python does with this.

*On the first line:* As always, assignment starts on the right.  So, Python first creates an object for `5`. Then, `var` is bound to that object: it is added to the table of variables, matched with the address of the object.



![IMAGE NOT FOUND!!!!!!!!!!!!](frame1.jpg)

*On the second line:* An object for `1` is created.  Then, Python looks to address `0x1` to retrieve the value of `var`, and then the values of these two objects are added, to create a new object, with value `6`.  That new object is bound to the variable `x`.


![IMAGE NOT FOUND!!!!!!!!!!!!](frame2.jpg)


<br><br><br><br><br><br><br><br><br><br>

*On the third line:* Finally, the last assignment doesn't involve any new objects being created.  Python checks which object is attached to `x`, and then that object is bound to `var`.  So now, both `var` and `x` look to address `0x3` for their value.


![IMAGE NOT FOUND!!!!!!!!!!!!](frame3.jpg)

*On the last line:* Since `var` is attached to `0x3`, Python looks at address `0x3` to retrieve what it will be printing out.

<br><br><br><br><br><br><br><br><br><br>

So far, none of this is particularly illuminating.  But when you bring in mutable objects like lists, things get more interesting.

```python
# EXAMPLE 5c: Lists

my_list = [8,9]
my_list[0] = 20
```

This code snippet plays out as follows:

*On the first line*: first, objects are created for the literals `8` and `9`. Then, a list object is created. **A list object contains the addresses of the values it contains.** And the list object is bound to `my_list`.


![IMAGE NOT FOUND!!!!!!!!!!!!](frame4.jpg)


*On the second line*: an object is created for the value `20`, and the address of that object is placed as the first entry of the list object at `0x13`. Let's emphasize this: **this line is a mutation, and so it doesn't change which object `my_list` is bound to -- it just alters that object**. 


![IMAGE NOT FOUND!!!!!!!!!!!!](frame5.jpg)


<br><br><br><br><br><br><br><br><br><br>

Finally, we can solve the riddle.

```python
# EXAMPLE 5d: The weird code.

x = ["Old", "One"]  # Line 1a
y = x                # Line 2a
x = ["New", "Guy"]  # Line 3a

x = ["Old", "One"]  # Line 1b
y = x                # Line 2b
x[0] = "New!!!!"     # Line 3b
```

Line 1a/b: Creates objects for `"Old"`, `"One"`, and the list; puts the addresses of the strings into the list; and binds `x` to the list object.


Line 2a/b: `x` is bound to the object at `0x23`, so this object now gets bound to `y` as well.


![IMAGE NOT FOUND!!!!!!!!!!!!](frame6.jpg)


*Now, here's the difference*

Line 3a: Creates objects for `"New"`, `"Guy"`, and a new list object to hold them; puts the addresses of the strings into the list; and binds `x` to the new list object.  **The old object at `0x23` isn't touched, and `y` is still bound it it.**


![IMAGE NOT FOUND!!!!!!!!!!!!](frame7.jpg)

Line 3b: Modifies the contents of the list object at `0x23`.  Note that neither of the variable bindings change -- they both still point to `0x23`, which has now been changed!
